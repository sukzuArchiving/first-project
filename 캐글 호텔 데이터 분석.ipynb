# 시간 시각화

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime

df=pd.read_csv("C:/Users/Administrator/Desktop/캐글 실습데이터/train.csv")
```

```python
df.head()
```

```python
df['data2'] = pd.to_datetime(df['Order Date'], infer_datetime_format=True)
```

```python
df = df.sort_values(by="data2")
```

```python
df
```

```python
df['year']=df['data2'].dt.year
```

```python
df_2018= df.loc[df.year==2018]
```

```python
df_2018.tail()
```

```python
# 2018 일별 매출액

df_2018 = df_2018.groupby('data2')['Sales'].sum().reset_index()
df_2018.head()
```

```python
import seaborn as sns
# 30일 이평선
df_2018['Month'] = df_2018['Sales'].rolling(window=20).mean()

# 선그래프

sns.lineplot(x='data2',y='Sales',data=df_2018)
sns.lineplot(x='data2',y='Month',data=df_2018)
```

```python
df_year_sale = df.groupby('year')['Sales',].sum().reset_index()

```

```python
df_year_sale.head()
```

```python
sns.barplot(x='year',y='Sales',data=df_year_sale)
```

```python
df.head()
```

```python
df_bar2 = df.groupby(['year','Segment'])['Sales'].sum().reset_index()
df_bar2.head()
```

```python
df_bar2_pv= df_bar2.pivot_table(index ="year",
                          columns='Segment',
                          values='Sales',
                          aggfunc='mean').reset_index()

df_bar2_pv.head()
```

```python
df_cumulative = df_bar2_pv.set_index('year').cumsum(axis=1).reset_index()
df_cumulative.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# df_bar2_pv를 year를 기준으로 누적해서 계산합니다.


# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 각 Segment에 대한 누적 막대 그래프 그리기
sns.barplot(x=df_cumulative['year'], y=df_cumulative['Consumer'], alpha=0.7)
sns.barplot(x=df_cumulative['year'], y=df_cumulative['Home Office'], alpha=0.7)
sns.barplot(x=df_cumulative['year'], y=df_cumulative['Corporate'], alpha=0.7)

# 그래프 제목 설정
plt.title('Cumulative Bar Chart of Sales by Year and Segment')

# x 축 레이블 설정
plt.xlabel('Year')

# y 축 레이블 설정
plt.ylabel('Cumulative Sales')

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
